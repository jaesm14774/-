```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import selenium.webdriver.support.ui as ui
import re
import time
import pickle
import os
```

```python
options = Options()
#options.add_argument('--headless')
#options.add_argument('--no-sandbox')
```

```python
browser=webdriver.Chrome('C:/Program Files/Anaconda3/chromedriver.exe',options=options)
```

```python
picture=[]
browser.get(url="https://tw.manhuagui.com/comic/19430/219425.html")
browser.implicitly_wait(2)
for i in range(0,176): #簡略寫法
    Temp=[]
    if i == 0: #不知為什麼，無法正確加載，先讓他正確加載
        browser.find_element_by_link_text('下一頁').click()
        browser.implicitly_wait(2)
        browser.find_element_by_link_text('上一頁').click()
        browser.implicitly_wait(18)
        time.sleep(3)
        #找到這回有幾頁
        soup=BeautifulSoup(browser.page_source,'lxml')
        page=re.search(string=str(soup.find(class_='w980 title')),pattern='>/\d{1,3}').group(0)
        page=int(re.sub(string=page,pattern='>/',repl=''))
        for j in range(0,page):
            soup=BeautifulSoup(browser.page_source,'lxml')
            Temp.append(soup.select('#mangaFile')[0].get('src'))
            if j is not (page-1):
                browser.find_element_by_link_text('下一頁').click()
                browser.implicitly_wait(3)
                time.sleep(2)
            else:
                time.sleep(10)
                print('Done'+str(i)+'回')
        picture.append(Temp)
        time.sleep(10)
        browser.find_element_by_link_text('下一章').click()
        browser.implicitly_wait(15)
        time.sleep(5)
    else:
        #找到這回有幾頁
        soup=BeautifulSoup(browser.page_source,'lxml')
        page=re.search(string=str(soup.find(class_='w980 title')),pattern='>/\d{1,3}').group(0)
        page=int(re.sub(string=page,pattern='>/',repl=''))
        for j in range(0,page):
            soup=BeautifulSoup(browser.page_source,'lxml')
            Temp.append(soup.select('#mangaFile')[0].get('src'))
            if j is not (page-1):
                browser.find_element_by_link_text('下一頁').click()
                browser.implicitly_wait(3)
                time.sleep(2)
            else:
                time.sleep(10)
                print('Done'+str(i)+'回')
        picture.append(Temp)
        time.sleep(3)
        if i is not 175:
            print(i)
            browser.find_element_by_link_text('下一章').click()
            browser.implicitly_wait(15)
            time.sleep(5)
        else:
            print('Done')
```
